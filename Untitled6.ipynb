```python
# ---------------------------------------
# Simple Library Management System
# using OOP , Lists , Functions , Lambda
# ---------------------------------------

# Book class represents a single book in the library
class Book:

    def __init__(self, title, author):
        self.title = title        # store book title
        self.author = author      # store author name
        self.available = True     # book availability

    def borrow_book(self):
        # check if the book is available
        if self.available:
            self.available = False
            print("Book borrowed successfully.")
        else:
            print("Sorry, this book is already borrowed.")

    def return_book(self):
        # make the book available again
        self.available = True
        print("Book returned successfully.")

    def show_info(self):
        # display book information
        print("\n--- Book Info ---")
        print("Title:", self.title)
        print("Author:", self.author)
        print("Available:", self.available)


# list to store all books
books = []


# function to add a new book
def add_book():

    print("\nAdd New Book")

    title = input("Enter book title: ")
    author = input("Enter author name: ")

    book = Book(title, author)

    books.append(book)

    print("Book added successfully.")


# function to show all books
def display_books():

    if len(books) == 0:
        print("\nNo books in the library.")
        return

    print("\n=== Library Books ===")

    for book in books:
        book.show_info()


# function to borrow a book
def borrow_book():

    title = input("\nEnter book title to borrow: ")

    for book in books:
        if book.title.lower() == title.lower():
            book.borrow_book()
            return

    print("Book not found.")


# function to return a book
def return_book():

    title = input("\nEnter book title to return: ")

    for book in books:
        if book.title.lower() == title.lower():
            book.return_book()
            return

    print("Book not found.")


# function to delete a book
def delete_book():

    title = input("\nEnter book title to delete: ")

    for book in books:
        if book.title.lower() == title.lower():
            books.remove(book)
            print("Book deleted successfully.")
            return

    print("Book not found.")


# Lambda example to get first book alphabetically
def show_first_book():

    if not books:
        print("\nNo books available.")
        return

    first_book = min(books, key=lambda b: b.title)

    print("\nFirst Book Alphabetically:")
    first_book.show_info()


# main menu
def main():

    while True:

        print("\n===== Library Management System =====")
        print("1. Add Book")
        print("2. Show All Books")
        print("3. Borrow Book")
        print("4. Return Book")
        print("5. Delete Book")
        print("6. Show First Book (Lambda)")
        print("7. Exit")

        choice = input("\nEnter your choice: ")

        if choice == "1":
            add_book()

        elif choice == "2":
            display_books()

        elif choice == "3":
            borrow_book()

        elif choice == "4":
            return_book()

        elif choice == "5":
            delete_book()

        elif choice == "6":
            show_first_book()

        elif choice == "7":
            print("\nExiting... Goodbye!")
            break

        else:
            print("\nInvalid choice. Try again.")


main()
```
